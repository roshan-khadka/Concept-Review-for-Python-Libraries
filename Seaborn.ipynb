## Getting Started with Seaborn

```python
# Statistical plotting library. Has beautiful default styles amd works beautifully with pandas objects
import seaborn as sns
```

```python
%matplotlib inline 
```

```python
# official dataset

tips = sns.load_dataset('tips')                     
flights = sns.load_dataset('flights')
tips.head()
```

```python
# Distribution plots

sns.distplot(tips["total_bill"], kde = False, bins = 30)
sns.kdeplot(tips['total_bill'])                                     # only kde plot

# Joint plot--comparison between two different distribution plot
sns.jointplot(x ='total_bill', y='tip', data = tips, kind ='resid')     # kind = reg, scatter, hex, kde density

# Pair plot-- joint plot for every possible(numerical) column in df
sns.pairplot(tips, hue = 'sex', palette ='coolwarm')

# Rug plot -- just uncounted distr plot
sns.rugplot(tips["total_bill"])
```

```python
# Categorical Plots -
# 1. Bar plot
sns.barplot(x='sex', y='total_bill', data=tips)         # avg of total bill compared to m or f

# 2. Countplot
sns.countplot(x='sex', data =tips)                      # simple count

# 3. Boxplot
sns.boxplot(x='day', y='total_bill',data=tips)

# 4. Violinplot                                         # more info than boxplot but also harder to read
sns.violinplot(x='day', y='total_bill',data=tips)

# 5. Strip Plot
sns.stripplot(x='day', y='total_bill', data=tips, jitter =True)

# 6. Swarm Plot
sns.swarmplot(x='day', y='total_bill', data=tips)

# 7. factor Plot
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')   # kind can be any from above plots
```

```python
# Matrix Plots - Need data in format of matrix(pivot or coorrelataion table)
tc = tips.corr()
sns.heatmap(tc, annot=True)

fp = flights.pivot(index ='month', columns ='year', values='passengers')
sns.heatmap(fp, linecolor ='black', linewidth =1)
```
